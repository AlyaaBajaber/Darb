```python
import random
import pandas as pd
from datetime import datetime, timedelta
from shapely.geometry import LineString, box
import folium
from tqdm import tqdm

def generate_weather():
    weather_conditions = ['Clear', 'Cloudy', 'Rainy', 'Stormy', 'Foggy']
    weights = [7, 4, 2, 1, 1]
    return random.choices(weather_conditions, weights=weights)[0]

def is_within_bounding_box(line, bbox):
    return bbox.intersects(line)

def generate_traffic_data(streets_gdf):
    traffic_data = []
    traffic_id = 0
    start_date = datetime(2023, 1, 1)
    end_date = datetime(2023, 6, 30)
    date_range = pd.date_range(start=start_date, end=end_date, freq='D')

    
    bbox1 = box(46.780923, 24.727429, 46.807406, 24.742193)
    bbox2 = box(46.672496, 24.674584, 46.683905, 24.677683)
    bbox3 = box(46.692658, 24.682768, 46.756313, 24.708039)

    for date in tqdm(date_range, desc="Processing Dates"):
        
        if date.weekday() == 4:  
            hours = range(18, 24)  #  6 PM to 12 AM
        elif date.weekday() == 5:  # Saturday
            hours = range(18, 24)  # 6 PM to 12 AM
        elif date.weekday() == 3:  # Thursday
            hours = range(6, 24)  #  6 AM to 12 AM
        else:
            hours = range(6, 19)  

        for hour in hours:
            datetime_instance = datetime.combine(date, datetime.min.time()) + timedelta(hours=hour)
            
           
            day_name = date.strftime('%A') 
            
            for _, street in streets_gdf.iterrows():
                street_name = street['name']
                line_geometry = street['geometry']
                weather = generate_weather()

               
                if is_within_bounding_box(line_geometry, bbox1) or is_within_bounding_box(line_geometry, bbox2):
                    traffic_status = 'Moderate'
                    num_cars = random.randint(20, 50)
                    avg_speed = round(random.uniform(30, 50), 2)
                elif is_within_bounding_box(line_geometry, bbox3):
                    traffic_status = 'Heavy'
                    num_cars = random.randint(50, 150)
                    avg_speed = round(random.uniform(40, 100), 2)
                else:
                    if 6 < datetime_instance.hour < 19:
                        num_cars = random.randint(30, 120)
                        avg_speed = round(random.uniform(5, 80), 2)
                    else:
                        num_cars = random.randint(10, 50)
                        avg_speed = round(random.uniform(40, 120), 2)

                    traffic_status = 'Heavy' if num_cars >= 50 and avg_speed <= 30 else 'Moderate' if num_cars < 30 else 'Light'

               
                traffic_data.append({
                    'traffic_id': traffic_id,
                    'time_series': datetime_instance,
                    'street_name': street_name,
                    'line_geometry': line_geometry,
                    'weather': weather,
                    'number_of_cars': num_cars,
                    'average_car_speed': avg_speed,
                    'traffic_status': traffic_status,
                    'day_name': day_name
                })
                traffic_id += 1

    return pd.DataFrame(traffic_data)


traffic_df = generate_traffic_data(khurais)


```

```python
import pandas as pd

def update_traffic_status(row):
    if row['average_car_speed'] > 90:
        return 'Light'
    elif 70 < row['average_car_speed'] <= 90:
        return 'Moderate'
    elif 40 < row['average_car_speed'] <= 70:
        return 'Moderate'
    elif row['average_car_speed'] <= 40:
        return 'Heavy'




traffic_df['traffic_status'] = traffic_df.apply(update_traffic_status, axis=1)
print(f"Updated shape: {traffic_df.shape}")
print(traffic_df[['average_car_speed', 'number_of_cars', 'traffic_status']].head(10))
```

```python
import numpy as np

average_passed_cars = 1000  

def update_number_of_cars(row):
    if row['traffic_status'] == 'Light':
        
        return np.random.randint(average_passed_cars + 100, average_passed_cars + 500)
    elif row['traffic_status'] == 'Moderate':
        
        return np.random.randint(average_passed_cars - 200, average_passed_cars + 200)
    elif row['traffic_status'] == 'Heavy':
        
        return np.random.randint(200, average_passed_cars - 100)
    else:
        
        return average_passed_cars  


traffic_df['number_of_cars'] = traffic_df.apply(update_number_of_cars, axis=1)
```

# generate accidents

```python
import random
import pandas as pd
from datetime import datetime, timedelta
from shapely.geometry import LineString, Point
import folium
from tqdm import tqdm 

def generate_accident_reason():
    reasons = ['Using Phone', 'Speeding', 'Reckless Driving', 'Road Conditions', 'Weather']
    weights = [7, 3, 4, 11, 1]
    return random.choices(reasons, weights=weights)[0]

def generate_accident_data(traffic_df):
    accidents_data = []

    
    traffic_df['date'] = traffic_df['time_series'].dt.date
    daily_traffic = traffic_df.groupby('date').size().reset_index(name='traffic_count')

    for _, daily_record in tqdm(daily_traffic.iterrows(), total=daily_traffic.shape[0], desc="Generating Accidents"):
        day = daily_record['date']
        daily_records = traffic_df[traffic_df['date'] == day]

        
        total_accidents = random.randint(2, 10)

        for _ in range(total_accidents):
            
            traffic_status_choice = random.choices(
                ['Heavy', 'Moderate', 'Light'], 
                weights=[60, 30, 10]
            )[0]

            
            status_records = daily_records[daily_records['traffic_status'] == traffic_status_choice]

            if not status_records.empty:
                traffic_record = status_records.sample(n=1).iloc[0]
                street_name = traffic_record['street_name']
                line_geometry = traffic_record['line_geometry']
                date_with_density = traffic_record['time_series']

                if isinstance(line_geometry, LineString):
                    line_coords = list(line_geometry.coords)

                    if len(line_coords) > 2:  
                        
                        point_index = random.randint(1, len(line_coords) - 2)
                        point_geom = Point(line_coords[point_index])
                    else:
                        
                        point_geom = Point(line_coords[0])

                    num_cars_involved = random.choices([1, 2, 3, random.randint(4, 10)], weights=[10, 18, 8, 1])[0]
                    accident_reason = generate_accident_reason()
                    num_injured = random.choices([0, random.randint(1, 5)], weights=[20, 5])[0]

                    accidents_data.append({
                        'accident_id': len(accidents_data),
                        'point_geometry': point_geom,
                        'street_name': street_name,
                        'number_of_cars_involved': num_cars_involved,
                        'time_series': date_with_density,
                        'accident_reason': accident_reason,
                        'number_of_injured': num_injured,
                    })

    return pd.DataFrame(accidents_data)

accident_df = generate_accident_data(traffic_df)

m = folium.Map(location=[24.7136, 46.6753], zoom_start=12)

for _, row in tqdm(traffic_df.iterrows(), total=traffic_df.shape[0], desc="Adding Traffic Data to Map"):
    line = row['line_geometry']
    if isinstance(line, LineString):
        coords = [(y, x) for x, y in line.coords]
        color = 'blue' if row['traffic_status'] == 'Light' else 'orange' if row['traffic_status'] == 'Moderate' else 'red'
        folium.PolyLine(coords, color=color, weight=2.5, opacity=0.7,
                        tooltip=f"{row['street_name']} - Cars: {row['number_of_cars']}, Speed: {row['average_car_speed']} km/h, Status: {row['traffic_status']}").add_to(m)


for _, accident in tqdm(accident_df.iterrows(), total=accident_df.shape[0], desc="Adding Accident Data to Map"):
    folium.Marker(
        location=[accident['point_geometry'].y, accident['point_geometry'].x],
        popup=f"Accident: {accident['accident_reason']} | Cars Involved: {accident['number_of_cars_involved']} | Injured: {accident['number_of_injured']}",
        icon=folium.Icon(color='red')
    ).add_to(m)
```

# convert to geo dataframe and Combined the traffics and accidents

```python
import geopandas as gpd

traffic_gdf = gpd.GeoDataFrame(traffic_df, geometry='line_geometry', crs='EPSG:4326')
accident_gdf = gpd.GeoDataFrame(accident_df, geometry='point_geometry', crs='EPSG:4326')

merged_df = gpd.sjoin(accident_gdf, traffic_gdf, how='inner', predicate='intersects')

merged_df = merged_df[merged_df['time_series_left'] == merged_df['time_series_right']]

merged_df = merged_df.drop_duplicates(subset=['accident_id', 'time_series_left'])

merged_df = merged_df.merge(traffic_gdf[['traffic_id', 'line_geometry']], on='traffic_id', how='left')

merged_df = merged_df[['accident_id', 'point_geometry', 'street_name_left', 
                        'number_of_cars_involved', 'time_series_left', 
                        'accident_reason', 'number_of_injured', 
                        'traffic_id', 'time_series_right', 
                        'street_name_right', 'weather', 
                        'number_of_cars', 'average_car_speed', 
                        'traffic_status', 'day_name', 'date', 
                        'line_geometry']]


print(f"Shape after removing duplicates: {merged_df.shape}")
print(merged_df.head(10))
```
